5. Generate a Synthetic Dataset with Correlated Features and Visualize the Correlation Matrix Using a Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(42)
data = np.random.multivariate_normal([0, 0, 0], [[1, 0.8, 0.6], [0.8, 1, 0.7], [0.6, 0.7, 1]], 100)
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3'])

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix Heatmap')

plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

np.random.seed(30)

data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

fig = go.Figure(data=[go.Scatter3d(
    x=df['X'],
    y=df['Y'],
    z=df['Z'],
    mode='markers',
    marker=dict(
        size=5,
        color='red',  
        opacity=0.8
    )
)])

fig.update_layout(
    scene=dict(
        xaxis_title='X Label',
        yaxis_title='Y Label',
        zaxis_title='Z Label'
    ),
    title='3D Scatter Plot'
)

fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Seed for reproducibility
np.random.seed(15)

# Generate the dataset
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a violin plot using Plotly Express
fig = px.violin(
    df, y='Score', x='Grade',
    box=True,  # Add a box plot inside the violin plot
    points='all',  # Show all data points
    title='Distribution of Scores Across Different Grade Categories',
    labels={'Score': 'Score', 'Grade': 'Grade'},
    height=600
)

# Show the plot
fig.show()
```

3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and 
days.


          np.random.seed(20)
          data = {
          'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 
            'May'], 100),
      
           'Day': np.random.choice(range(1, 31), 100),
           'Sales': np.random.randint(1000, 5000, 100)
           }
 
           df = pd.DataFrame(data)

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

fig = px.imshow(
    heatmap_data,
    title='Heatmap of Sales Across Different Months and Days',
    labels={'x': 'Month', 'y': 'Day', 'color': 'Sales'},
    color_continuous_scale='Viridis'
)

fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(
    z=z,
    x=x,
    y=y,
    colorscale='Viridis'
)])

fig.update_layout(
    title='3D Surface Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

fig.show()
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_file

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, legend_label="Sine Wave", line_width=2)

output_file("sine_wave.html")

show(p)


# file:///C:/Users/AMAN/OneDrive/Desktop/sine_wave.html
```

2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.io import show
from bokeh.transform import factor_cmap

np.random.seed(42)
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
sizes = np.random.uniform(10, 50, 100)  
colors = np.random.choice(['red', 'green', 'blue', 'orange'], 100)  

df = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label='x', y_axis_label='y')

color_mapper = factor_cmap('colors', palette=['red', 'green', 'blue', 'orange'], factors=['red', 'green', 'blue', 'orange'])

p.scatter(
    x='x',
    y='y',
    size='sizes',
    color=color_mapper,
    legend_field='colors',
    source=df,
    fill_alpha=0.6
)

output_file("scatter_plot.html")

show(p)
# file:///C:/Users/AMAN/OneDrive/Desktop/scatter_plot.html
```

 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.


         fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] 
         counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show, output_file
from bokeh.io import show
from bokeh.palettes import Category10

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

data = {'fruits': fruits, 'counts': counts}

p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruit', y_axis_label='Count',
           toolbar_location=None, tools="")

p.vbar(x=data['fruits'], top=data['counts'], width=0.4, color=Category10[len(fruits)])

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.axis.minor_tick_line_color = None
p.outline_line_color = None

output_file("bar_chart.html")

show(p)

# file:///C:/Users/AMAN/OneDrive/Desktop/bar_chart.html
```

 4. Create a Bokeh histogram to visualize the distribution of the given data.
 
         data_hist = np.random.randn(1000)

```python

import numpy as np
from bokeh.plotting import figure, show, output_file

np.random.seed(42)  
data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Normally Distributed Data", x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

output_file("histogram.html")

show(p)

# file:///C:/Users/AMAN/OneDrive/Desktop/histogram.html
```

5. Create a Bokeh heatmap using the provided dataset.
 data_heatmap = np.random.rand(10, 10)
 
            x = np.linspace(0, 1, 10)
            y = np.linspace(0, 1, 10)
            xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

np.random.seed(42)  
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', 
           x_range=(0, 1), y_range=(0, 1),
           tools="hover,save,pan,box_zoom,reset,wheel_zoom",
           tooltips=[('X', '@x'), ('Y', '@y'), ('Value', '@z')],
           )

mapper = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

p.rect(x='x', y='y', width=0.1, height=0.1, source={'x': x_flat, 'y': y_flat, 'z': z_flat},
       fill_color=mapper, line_color=None)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

output_file("heatmap.html")

show(p)

# file:///C:/Users/AMAN/OneDrive/Desktop/heatmap.html
```
